# My first project

## Load data


```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

# Data Preparation


## Data seperation as x and y

```python
y= df["logS"]
y
```

```python
X=df.drop("logS",axis=1)# axis=1 allows to work with columns and axis=0 rows 
X
```

## Data Splitting 


```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

# Model Building

## Linear Regression



```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

### Applying model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)#making predictions on trained dataset

y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

## Evaluate Model Performance

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)

lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)

lr_test_r2 = r2_score(y_test,y_lr_test_pred)

print("this is your lr_train MSE score:", lr_train_mse)
print("this is your lr_train r2 score:", lr_train_r2)
print("this is your lr_test MSE score:", lr_test_mse)
print("this is your lr_test r2 score:", lr_test_r2)
```

```python
lr_results = pd.DataFrame  (["Linear regression",lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
#transpose turns the rows to columns and columns to rows like in linear algebra
```

```python
lr_results.columns=["Method","Training MSE","Training R2","Test MSE","Test R2"]
lr_results
```

## Random Forest

### Training the model

```python
from sklearn.ensemble import RandomForestRegressor #using regressor because y variable quantative so we build regression
rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train,y_train)
```

### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)

y_rf_test_pred = rf.predict(X_test)
```

### Evaluate model performace

```python
rf_train_mse = mean_squared_error(y_train , y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)


rf_test_mse = mean_squared_error(y_test , y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

print("this is your rf_train MSE score:", rf_train_mse)
print("this is your rf_train r2 score:", rf_train_r2)
print("this is your rf_test MSE score:", rf_test_mse)
print("this is your rf_test r2 score:", rf_test_r2)
```

```python
rf_results = pd.DataFrame(["Random Forest",rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()

rf_results.columns = ["Method","Training MSE","Training R2","Test MSE","Test R2"]
```

```python
rf_results
```

## Model comparison

```python
df_models = pd.concat([lr_results,rf_results],axis=0)#axis = 0 makes them into rows which is what we want
```

```python
df_models # the closer mse to 0 the better 
```
