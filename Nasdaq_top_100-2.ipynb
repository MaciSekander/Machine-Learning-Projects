```python
import pandas as pd 
import yfinance as yf 
import numpy as np
```

```python
ticker_df = pd.read_html('https://en.wikipedia.org/wiki/Nasdaq-100')[4]#using pandas hyml function
ticker_df
```

```python
tickers = ticker_df.Ticker.to_list()# converting the Ticker column to a list
tickers
```

```python
df = yf.download(tickers,start='2010-01-01')['Adj Close']
#here the NaN values are companies that might have been traded at that time on the NASDAQ
#so we have a survivorship bias 
```

```python
df
```

```python
df = df.dropna(axis=1)
```

```python
mtl = (df.pct_change()+1)[1:].resample('M').prod()#this is resampling the data monthly base and taking the product(monthly returns)
#getting the cumulative returns
```

```python
mtl #this is monthly returns by using the percent change by month 
# so if you did this by hand the apple price on jan31/jan1
```

```python
#know we want 3 month,6month,12month returns by making a function
def get_rolling_ret(df,n):# n for the month we want the cumulative return for the certain time period
    return df.rolling(n).apply(np.prod)
    
```

```python
get_rolling_ret(mtl,12).head(12)#we get the 12 month return
#first 11 months will be NaN but the last rows will give us our 12 month returns
```

```python
ret_12,ret_6,ret_3 = get_rolling_ret(mtl,12),get_rolling_ret(mtl,6),get_rolling_ret(mtl,3)
#doing this for efficency so don't have to keep calculating
```

```python
ret_6#rolling returns over 6 months so the performance 
```

```python
top_50 = ret_12.loc['2010-12-31'].nlargest(50).index
top_50#showing our top 50 [performers in this time period
```

```python
top_30 = ret_6.loc['2010-12-31',top_50].nlargest(30).index
top_30
```

```python
top_10 = ret_3.loc['2010-12-31',top_30].nlargest(10).index
top_10 #to see the returns just remove index
```

```python
def get_top(date):
    top_50 = ret_12.loc[date].nlargest(50).index
    top_30 = ret_6.loc[date,top_50].nlargest(30).index
    top_10 = ret_3.loc[date,top_30].nlargest(10).index
    return top_10
```

```python
get_top('2010-12-31')
```

```python
portfolio = mtl.loc['2010-12-31':,get_top('2010-12-31')][1:2]#slicing monthly to get the next date
portfolio
```

```python
portfolio.mean(axis=1)#.values[0] #mean of the your portfolios columns
#portfolio return of roughly 5% on that given day
```

```python
def pf_performance(date):
    portfolio = mtl.loc[date:,get_top(date)][1:2]
    return portfolio.mean(axis=1).values[0]
```

```python
pf_performance('2010-12-31')
#portfolio performance of 5% on this given day
```

```python
mtl.index[:-1] #getting our dates
```

```python
mtl.index[1:]
```

```python
returns = []

for date in mtl.index[:-1]:
    returns.append(pf_performance(date))
```

```python
pd.Series(returns,index=mtl.index[1:]).cumprod() #taking the cumulative product and plot
```

```python
pd.Series(returns,index=mtl.index[1:]).cumprod().plot() #taking the cumulative product and plot
#equity curve
```

```python
pd.Series([i - 0.01 for i in returns],index=mtl.index[1:]).cumprod().plot() #taking the cumulative product and plot
#with 1% commtion fees
```
